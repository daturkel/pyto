## Optimizing Next Steps

### Truth Booth

Truth Booth strategy is simple, the best couple to send is always the one with the odds closest to 50%, since this guarantees the greatest expected reduction in remaining scenarios.

Since Jasmine and Nour have a 41.8% chance of being a Perfect Match, they should be the couple sent to the Truth Booth.

### Matchup Ceremony

Supposing we skipped the Truth Booth, what would be the best set of matches for the next ceremony?

```python
analysis = season_8.analyze_scenarios()
```

```python
argmax_min = []
max_min = 0
argmax_median = []
max_median = 0
argmax_mean = []
max_mean = 0
argmax_mode = []
max_mode = 0
for scenario, data in analysis.items():
    if data["min_beams"] > max_min:
        max_min = data["min_beams"]
        argmax_min = [scenario]
    elif data["min_beams"] == max_min:
        argmax_min.append(scenario)
    if data["mean_beams"] > max_mean:
        max_mean = data["mean_beams"]
        argmax_mean = [scenario]
    elif data["mean_beams"] == max_mean:
        argmax_mean.append(scenario)
    if data["median_beams"] > max_median:
        max_median = data["median_beams"]
        argmax_median = [scenario]
    elif data["median_beams"] == max_median:
        argmax_median.append(scenario)
    try:
        if data["mode_beams"] > max_mode:
            max_mode = data["mode_beams"]
            argmax_mode = [scenario]
        elif data["mode_beams"] == max_mode:
            argmax_mode.append(scenario)
    except TypeError:
        pass

print(f"max_min beams={max_min}")
print(f"{len(argmax_min)} such scenarios")
print()
print(f"max_median beams={max_median}")
print(f"{len(argmax_median)} such scenarios")
print()
print(f"max_mean beams={max_mean}")
print(f"{len(argmax_mean)} such scenarios")
print()
print(f"max_mode beams={max_mode}")
print(f"{len(argmax_mode)} such scenarios")
```
