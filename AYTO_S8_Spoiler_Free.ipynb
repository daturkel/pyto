## Optimizing Next Steps

### Truth Booth

Truth Booth strategy is simple, the best couple to send is always the one with the odds closest to 50%, since this guarantees the greatest expected reduction in remaining scenarios.

There is a couple (redacted) who have a 41.8% chance of being a Perfect Match, so they should be the couple sent to the Truth Booth.

### Matchup Ceremony

Supposing we skipped the Truth Booth, what would be the best set of matches for the next ceremony?

There are multiple potential optimization problems.

```python
beam_results, beam_results_metadata, scenario_metadata = beam_analysis(season_8.remaining_scenarios)
```

Unfortunately no scenarios provide maximal-mean-beams *and* minimal max remaining scenarios.

```python
set(beam_results_metadata['max_mean_beams']['arg']) & set(scenario_metadata['min_max_scenarios']['arg'])
```

Of the scenarios providing max-mean-beams, we could have worst case 21 or 26 scenarios left.

```python
for guess in beam_results_metadata['max_mean_beams']['arg']:
    print(guess)
    print(beam_results[guess]['max_scenarios'])
    print()
```

Of the scenarios with minimal max-scenarios-remaining, the higest average number expected number of beams is: 3.25.

```python
for guess in scenario_metadata['min_max_scenarios']['arg']:
    print(guess)
    print(beam_results[guess]['mean_beams'])
    print()
```
